```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
import pickle
```

```python
past_mean = pickle.load(open('./past_mean.pkl','rb'))
past_abs_sum = pickle.load(open('./past_abs_sum.pkl','rb'))
past_median = pickle.load(open('./past_median.pkl','rb'))
past_lower_quartile = pickle.load(open('./past_lower_quartile.pkl','rb'))
past_upper_quartile = pickle.load(open('./past_upper_quartile.pkl','rb'))
past_std = pickle.load(open('./past_std.pkl','rb'))

ls_past = [past_mean, past_abs_sum, past_median, past_lower_quartile, past_upper_quartile, past_std]
ls_title = ['past_mean', 'past_abs_sum', 'past_median', 'past_lower_quartile', 'past_upper_quartile', 'past_std']
```

```python
def plot_data(sample_tensors, title):
    # We'll use a loop to plot each tensor's values with the specified colors
    plt.figure(figsize=(14, 8))

    for epoch, tensor in enumerate(sample_tensors, start=1):
        # Extract values for easier handling
        values = tensor.cpu().numpy().flatten()  # Convert each tensor from [50, 1] to [50]
        
        # Plot first 10 values in red
        plt.scatter([epoch] * 10, values[:10], color='red', s=10)
        
        # Plot remaining values in green
        plt.scatter([epoch] * 40, values[10:], color='green', s=10)

    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel('Value')
    plt.show()
```

```python
for i in range(len(ls_past)):
    plot_data(ls_past[i], ls_title[i])
```
